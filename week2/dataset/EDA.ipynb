## Combining all datasets into single

```python
# Population and Users columns will change into numerical
final_data['Population'] = final_data.Population.str.replace(',','').astype(float)
final_data['Users'] = final_data.Users.str.replace(',','').astype(float)
```

```python
print(final_data['Population'].dtype)
print(final_data['Users'].dtype)
```

```python
# converting date os travel to dates
min_date = final_data['Date of Travel'].agg('min')
max_date = final_data['Date of Travel'].agg('max')
print(min_date)
print(max_date)
max_date-min_date
```

```python
current_date = pd.to_datetime('2016-01-31')
int(current_date.strftime("%Y%m%d"))
```

```python
base_date = datetime.date.fromordinal(datetime.datetime(2016,1,31).toordinal()-min_date)
base_date
```

```python
final_data['Dates']= pd.to_datetime(final_data['Date of Travel'], unit='D', origin=pd.Timestamp('1900-01-28'))
final_data
```

```python
# Company profits
final_data['Profit'] = final_data['Price Charged'] - final_data['Cost of Trip']
final_data
```

```python
df = final_data.drop(['Date of Travel'], axis=1)
df.head()
```

```python
df.columns
```

```python
# According graph yellow cab making more money
sns.barplot(x='Company', y='Profit', data=df)
```

```python
# Total profit for each company
total_profit = df.groupby('Company')['Profit'].sum()
total_profit
```

```python
#Total km driven by each company
total_km = df.groupby('Company')['KM Travelled'].sum()
total_km
```

```python
#Total number of rides for each company
number_of_rides = df['Company'].value_counts()
number_of_rides
```

```python
#Profit that each company making from per ride
profit_per_ride = total_profit/number_of_rides
profit_per_ride
```

```python
# Yellow Cab making almost 2.5 time more money than Pink Cab for per ride

plt.bar(['Pink', 'Yellow'], profit_per_ride, color = ['blue', 'orange'])
plt.title('Profit for per ride')
```

```python
# Profit that each company making from per km
profit_per_km = total_profit/total_km
profit_per_km
```

```python
# Yellow Cab making almost 2.5 time more money than Pink Cab for per km
plt.bar(['Pink', 'Yellow'], profit_per_km, color = ['blue', 'orange'])
plt.title('Profit for per km')
```

```python
df['year'] = df['Dates'].dt.year
df
```

```python
# Average profit for per km in different years 
profit_by_year = df.groupby(['Company', 'year'])['Profit'].mean()
km_driven  = df.groupby(['Company', 'year'])['KM Travelled'].mean()
avg_profit_by_year = profit_by_year/km_driven
avg_profit_by_year
```

```python
# Total profit for each company by given years 
total_profit_by_year = df.groupby(['Company', 'year'])['Profit'].size().reset_index()
total_profit_by_year
```

```python
profit_by_gender = df.groupby(['year', 'Gender','Company'])['Profit'].size().reset_index()
profit_by_gender
```

```python
sns.barplot(x='year', y='Profit', data=profit_by_gender)
```
